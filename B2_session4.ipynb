```python
import numpy as np

marks = np.array([78, 91, 55, 72, 40, 88, 98, 81, 11, 23, 78, 88, 91, 96])

# 1. First the class average
average = np.mean(marks)
print(f"Class Average: {average:.2f}")

# 2. Find the highest score
highest = np.max(marks)
print(f"Highest Score: {highest}")

# 3. Find the lowest score
lowest = np.min(marks)
print(f"Lowest Score: {lowest}")

# 4. Count students above 75
count_above_75 = np.sum(marks > 75)
print(f"Students above 75: {count_above_75}")
```

```python
import pandas as pd
#STUDENT PERFORMANCE ANALYSER

# Read session4_dummy_students.csv.

df = pd.read_csv("session4_dummy_students_data.csv")
print(df)

# Create an average column using math, science, and english.

df['average'] = (df['math'] + df['science'] + df['english']) / 3
print(df[['name', 'average']])
```

```python
# 4. Print class average
class_avg = df['average'].mean()
print(f"Class Average: {class_avg:.2f}\n")
```

```python
# 5. Print subject wise averages
print("Subject-wise Averages:")
print(f" - Science: {df['science'].mean():.2f}")
print(f" - English: {df['english'].mean():.2f}")
print(f" - Math: {df['math'].mean():.2f}\n")
```

```python
# 6. Find and print the top student(s)
highest_score = df['average'].max()
top_students = df[df['average'] == highest_score]
print("Top Student(s):")
print(top_students[['name', 'average']])
print()
```

```python
# 7. Print students with attendance below 75
low_attendance = df[df['attendance_percent'] < 75]
print("Students with attendance below 75%:")
print(low_attendance[['name', 'attendance_percent']])
print()
```

```python
# 8. Print students who submitted fewer than 6 assignments
few_assignments = df[df['assignments_submitted'] < 6]
print("Students who submitted fewer than 6 assignments:")
print(few_assignments[['name', 'assignments_submitted']])
```
